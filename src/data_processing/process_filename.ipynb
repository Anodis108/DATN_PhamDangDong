## Đổi tên theo Name

```python

import os

# ====== CẤU HÌNH ======
folder_a = '/home/anodi108/Desktop/project/Do_An_Tot_Nghiep/DATN_PhamDangDong/DATN_PhamDangDong/resource/data/data/Data_DATN/Data_loc'

# ====== ĐỔI TÊN ẢNH TRONG TỪNG FOLDER CON ======
for subfolder in os.listdir(folder_a):
    subfolder_path = os.path.join(folder_a, subfolder)
    if not os.path.isdir(subfolder_path):
        continue

    try:
        name, height = subfolder.split('_')
    except ValueError:
        print(f"❌ Bỏ qua thư mục không hợp lệ: {subfolder}")
        continue

    # ===== 1. Đổi tên tạm thời để tránh ghi đè =====
    for file_name in os.listdir(subfolder_path):
        ext = os.path.splitext(file_name)[-1].lower()
        if ext not in ['.jpg', '.jpeg', '.png']:
            continue
        old_path = os.path.join(subfolder_path, file_name)
        temp_name = f"temp_{file_name}"
        temp_path = os.path.join(subfolder_path, temp_name)
        os.rename(old_path, temp_path)

    # ===== 2. Đổi từ tên tạm sang tên chuẩn =====
    idx = 1
    for file_name in sorted(os.listdir(subfolder_path)):
        if not file_name.startswith("temp_"):
            continue
        ext = os.path.splitext(file_name)[-1].lower()
        new_name = f"{name}_{idx}_{height}{ext}"
        src_path = os.path.join(subfolder_path, file_name)
        dst_path = os.path.join(subfolder_path, new_name)
        os.rename(src_path, dst_path)
        print(f"✅ Đổi tên {file_name} → {new_name}")
        idx += 1

print("🎉 Đã đổi tên tất cả ảnh trong các thư mục con.")
```

```python
# Process img name and rotate if needed
import os
import cv2
import numpy as np
import shutil

# ====== 1. CẤU HÌNH ======
folder_a = '/home/anodi108/Desktop/project/Do_An_Tot_Nghiep/DATN_PhamDangDong/DATN_PhamDangDong/resource/data/data/Data_DATN/Data_loc'
folder_b_parent = '/home/anodi108/Desktop/project/Do_An_Tot_Nghiep/DATN_PhamDangDong/DATN_PhamDangDong/resource/data/data/Data_DATN/data_chia'
output_dir = '/home/anodi108/Desktop/project/Do_An_Tot_Nghiep/DATN_PhamDangDong/DATN_PhamDangDong/resource/data/data/processed_data'
os.makedirs(output_dir, exist_ok=True)

# ====== 2. DUYỆT TỪNG FOLDER B ======
for b_folder in os.listdir(folder_b_parent):
    b_folder_path = os.path.join(folder_b_parent, b_folder)
    if not os.path.isdir(b_folder_path):
        continue

    prefix = b_folder.replace('.', '_')  # VD: "1.B" → "1_B"

    for b_file in os.listdir(b_folder_path):
        b_path = os.path.join(b_folder_path, b_file)
        try:
            img_b = cv2.imread(b_path)
            if img_b is None:
                print(f"⚠️ Không thể đọc ảnh {b_path}")
                continue
        except:
            continue

        matched = False

        # Duyệt ảnh trong mọi folder con của folder_a
        for sub_dir in os.listdir(folder_a):
            sub_dir_path = os.path.join(folder_a, sub_dir)
            if not os.path.isdir(sub_dir_path):
                continue

            for a_file in os.listdir(sub_dir_path):
                a_path = os.path.join(sub_dir_path, a_file)
                try:
                    img_a = cv2.imread(a_path)
                    if img_a is None:
                        continue

                    # So sánh nội dung ảnh
                    if img_a.shape == img_b.shape and np.array_equal(img_a, img_b):
                        new_name = f"{prefix}_{a_file}"
                        dst_path = os.path.join(output_dir, new_name)

                        # ==== 3. Nếu prefix là 5_ hoặc 6_, xoay ảnh trước khi lưu ====
                        if prefix.startswith("5") or prefix.startswith("6"):
                            rotated_img = cv2.rotate(img_a, cv2.ROTATE_90_CLOCKWISE)
                            cv2.imwrite(dst_path, rotated_img)
                        else:
                            shutil.copyfile(a_path, dst_path)

                        print(f"✅ Trùng khớp:")
                        print(f"   → A path: {a_path}")
                        print(f"   → B path: {b_path}")
                        print(f"   → Saved as: {dst_path}")
                        matched = True
                        break

                except Exception as e:
                    print(f"⚠️ Lỗi khi xử lý {a_path}: {e}")
                    continue

            if matched:
                break

        if not matched:
            print(f"❌ Không tìm thấy ảnh tương ứng với {b_file}")

print("🎉 Hoàn tất đối chiếu và sao chép.")
```

```python
# Xoay blackbox 90 độ để làm data
import os
import cv2
from ultralytics import YOLO

# ==== 1. CẤU HÌNH =====
model_path = '/home/anodi108/Desktop/project/Do_An_Tot_Nghiep/DATN_PhamDangDong/DATN_PhamDangDong/resource/weights/best_17052025_y11m_640.pt'  # thay đường dẫn model của bạn
input_folder = '/home/anodi108/Desktop/project/Do_An_Tot_Nghiep/DATN_PhamDangDong/DATN_PhamDangDong/resource/data/data/Data_DATN/data_chia/5_NamThang/process'
output_folder = '/home/anodi108/Desktop/project/Do_An_Tot_Nghiep/DATN_PhamDangDong/DATN_PhamDangDong/resource/data/data/Data_DATN/data_chia/5_NamThang/tesst_process'
os.makedirs(output_folder, exist_ok=True)

# ====== 2. Load YOLO Model ======
model = YOLO(model_path)

# ====== 3. Hàm xử lý ảnh ======
def rotate_and_replace(img, x1, y1, x2, y2, pad=10):
    h_img, w_img = img.shape[:2]
    cx = (x1 + x2) // 2
    cy = (y1 + y2) // 2
    bbox_w, bbox_h = x2 - x1, y2 - y1
    square_size = max(bbox_w, bbox_h) + pad

    # Tính toạ độ hình vuông dựa vào center
    x_start = max(cx - square_size // 2, 0)
    y_start = max(cy - square_size // 2, 0)
    x_end = min(x_start + square_size, w_img)
    y_end = min(y_start + square_size, h_img)

    # Điều chỉnh lại nếu bị tràn biên (cập nhật lại toạ độ bắt đầu)
    x_start = max(x_end - square_size, 0)
    y_start = max(y_end - square_size, 0)

    # Crop và xoay
    crop = img[y_start:y_end, x_start:x_end]
    rotated = cv2.rotate(crop, cv2.ROTATE_90_CLOCKWISE)

    # Paste lại hình xoay
    img[y_start:y_end, x_start:x_end] = rotated
    return img

# ====== 4. Duyệt qua từng ảnh và xử lý ======
for filename in os.listdir(input_folder):
    if not filename.lower().endswith(('.jpg', '.jpeg', '.png')):
        continue

    img_path = os.path.join(input_folder, filename)
    img = cv2.imread(img_path)
    if img is None:
        print(f"⚠️ Không đọc được ảnh: {img_path}")
        continue

    results = model(img)[0]
    for box in results.boxes:
        x1, y1, x2, y2 = map(int, box.xyxy[0])
        img = rotate_and_replace(img, x1, y1, x2, y2)

    out_path = os.path.join(output_folder, filename)
    cv2.imwrite(out_path, img)
    print(f"✅ Đã xử lý và lưu: {out_path}")

print("🎉 Hoàn tất xử lý toàn bộ ảnh.")
```

# Lọc ảnh trùng lặp file name


```python
import os
import cv2
import hashlib

# ===== 1. CẤU HÌNH =====
root_dir = '/home/anodi108/Desktop/project/Do_An_Tot_Nghiep/DATN_PhamDangDong/DATN_PhamDangDong/resource/data/data/Data_DATN/Data_loc'  # Thay đường dẫn folder A
valid_ext = ('.jpg', '.jpeg', '.png')

# ===== 2. Hàm tạo hash từ nội dung ảnh =====
def image_hash(image_path):
    try:
        img = cv2.imread(image_path)
        if img is None:
            return None
        img_bytes = cv2.imencode('.jpg', img)[1].tobytes()
        return hashlib.md5(img_bytes).hexdigest()
    except Exception as e:
        print(f"⚠️ Lỗi khi đọc ảnh {image_path}: {e}")
        return None

# ===== 3. So sánh và xóa ảnh trùng =====
hash_map = {}  # hash → path đầu tiên giữ lại

for subdir, _, files in os.walk(root_dir):
    for file in files:
        if not file.lower().endswith(valid_ext):
            continue

        file_path = os.path.join(subdir, file)
        h = image_hash(file_path)
        if h is None:
            continue

        if h in hash_map:
            # Nếu trùng hash với ảnh trước → xóa
            print(f"🗑️ Xoá {file_path} (trùng với {hash_map[h]})")
            os.remove(file_path)
        else:
            hash_map[h] = file_path  # Lưu hash ảnh đầu tiên

print("🎉 Đã xóa xong các ảnh trùng lặp.")
```
